# Selection Sort

Selection sort is an algorithm used for sorting the elements of an array by repeatedly finding the meinimum element from the unsorted array and placing it in front. There are two sub arrays maintained:

1. The sub array that has sorted elements.
2. Another sub array that is unsorted.

In each iteration an element with minimum value is picked from the unsorted array and placed in the sorted sub array of elements.

#### COMPLEXITY

- TIME COMPLEXITY
    - There are 2 loops so the complexity is n*n = n^2.
    - Worst Case Complexity: O(n2)
      If we want to sort in ascending order and the array is    in descending order then, the worst case occurs.
    - Best Case Complexity: O(n2)
       It occurs when the array is already sorted
    - Average Case Complexity: O(n2)
        It occurs when the elements of the array are in jumbled order (neither ascending nor descending).
- Space Complexity:
    Space complexity is O(1) because an extra variable 'i' is used.


#### ADVANTAGES

- The main advantage of the selection sort is that it performs well on a small list.
- Because it is an in-place sorting algorithm, no additional temporary storage is required beyond what is needed to hold the original list.
- Its performance is easily influenced by the initial ordering of the items before the sorting process.

#### DISADVANTAGES

- The primary disadvantage of the selection sort is its poor efficiency when dealing with a huge list of items.
- The selection sort requires n-squared number of steps for sorting n elements.
- Quick Sort is much more efficient than selection sort

#### ALGORITHM

- Set Min_index location to 0
- Traverse the list and find the minimum element of the list
- Swap it with value of minimum location
- Increment Min_index to point to next element
- Iterate till the list is sorted

#### HOW IT WORKS?

Let us assume an unsorted array with 4 elements in it.
arr[] = 25 11 80 22

Step 1. First of all find the minimum element in the array arr[0...3], in this example it is 11.

Step 2. Place it in the beginning of the array.
- OUTPUT 11 25 80 22

Now again traverse arr[1...3] to find the minimum element.  This time it is 22. So swap it with the frst element of arr[1...3].
- OUTPUT 11 22 80 25

Traverse now arr[2..3]. Swap the minimum element with the first element of arr[2..3]
- OUTPUT 11 22 25 80

Hence the array has been sorted.

#### Now let us see the execution of Selection Sort using Python

```python
def selectionSort(array, length):
   
    for step in range(length):
        min_idx = step

        for i in range(step + 1, length):
         
            # select the minimum element in each loop
            if array[i] < array[min_idx]:
                min_idx = i
         
        # put min at the correct position
        (array[step], array[min_idx]) = (array[min_idx], array[step])


data = [11, 5, 0, 56, 41]
length = len(data)
selectionSort(data, length)
print('Sorted Array in Ascending Order:')
print(data)
```
